###  Display sin value for the numbers in the interval [0,20]

```python
a=np.linspace(0,20,100,endpoint=True)
s,c=np.sin(a),np.cos(a)
plt.plot(a,s)
plt.plot(a,c)
plt.legend(['sine','cosine'],loc='upper left')
plt.show()
```

###  customizing

###  subplot

```python
fig=plt.figure()
fig.subplots_adjust(bottom= 0.025,left=0.025,top=0.975, right=0.975)

plt.subplot(2,3,1)
plt.bar(x,marks)
plt.xticks((np.linspace(0,125,3))), plt.yticks((np.linspace(0,90,3)))

plt.subplot(2,3,2)
plt.xticks(()), plt.yticks(())

plt.subplot(2,3,3)
plt.xticks(()), plt.yticks(())

plt.subplot(2,3,4)
plt.xticks(()), plt.yticks(())

plt.subplot(2,3,5)
plt.xticks(()), plt.yticks(())

plt.subplot(2,3,6)
plt.xticks(()), plt.yticks(())
```
